```python
#import library
import pandas as pd
```

```python
#load Data
df = pd.read_csv('data-indeks-standar-pencemar-udara-(ispu)-di-provinsi-dki-jakarta.csv')
df
```

```python
#drop column yang tidak dibutuhkan
df = df.drop(['periode_data', 'stasiun','parameter_pencemar_kritis','max'], axis = 1)
df.head()
```

```python
#drop baris yang tidak dibutuhkan
df = df[df['kategori'] != 'TIDAK ADA DATA']
```

```python
#cek data null
df.isnull().any()
df = df.dropna()
```

```python
columns_to_clean = ['pm_sepuluh', 'pm_duakomalima', 'sulfur_dioksida', 'karbon_monoksida', 'ozon', 'nitrogen_dioksida','kategori']
df[columns_to_clean] = df[columns_to_clean].replace(['-', '---'], 0)
```

```python
df = df.astype({'pm_sepuluh' : 'int','pm_duakomalima' : 'int','sulfur_dioksida' : 'int','karbon_monoksida' : 'int','ozon' : 'int','nitrogen_dioksida' : 'int'})
```

```python
df['tanggal'] = pd.to_datetime(df["tanggal"])
df.dtypes
```

```python
df = df.sort_values(by="tanggal", ascending=True, ignore_index=True)
```

```python
# Temukan nilai rata-rata pada setiap kolom numerik
numerical_columns = df.select_dtypes(include=['int']).columns
mean_values = df[numerical_columns].mean()

# Ganti nilai 0 dengan nilai rata-rata pada setiap kolom
df[numerical_columns] = df[numerical_columns].replace(0, mean_values)
df[numerical_columns] = df[numerical_columns].round(0).astype(int)
```

```python
#save data
df.to_csv("Data Pencemaran Udara Jakarta Cleaned.csv", index=False)
```
